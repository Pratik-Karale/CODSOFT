```python
import pandas as pd
```

```python
movies=None
movies=pd.read_csv("IMDb Movies India.csv",encoding='ISO-8859-1')

movies.index=range(movies.shape[0])
movies=movies.dropna(subset=['Rating','Genre','Actor 1','Actor 2','Actor 3'])

movies["Duration"]=movies["Duration"].str.extract("(\d+)").astype("Int16")
movies["Year"]=movies["Year"].str.extract("(\d+)").astype("Int16")
print(movies.shape)

movies.head()
# movies.Rating.describe()
movies.isnull().sum()
# movies.median()
# movies.dtypes
# movies.describe()
# movies["Actor 3"].nunique()
```

```python
movies.iloc[1]
```

```python
from sklearn.impute import SimpleImputer
mode_imputer=SimpleImputer(strategy="most_frequent")
movies[["Duration"]]=mode_imputer.fit_transform(movies[["Duration"]])
# movi
print(movies.Duration.isnull().sum())

movies.isnull().sum()
```

```python
movies.nunique()
```

```python
movies["Actor 1"] = movies.groupby('Actor 1')['Rating'].transform("mean")
movies["Actor 2"] = movies.groupby('Actor 2')['Rating'].transform("mean")
movies["Actor 3"] = movies.groupby('Actor 3')['Rating'].transform("mean")
movies["Director"] = movies.groupby('Director')['Rating'].transform("mean")
# temp_movies=movies
# # 2. Replace the original Genre column with these calculated means
# temp_movies['Actor 1'] = actor_means

# print(movies[['Actor 1', 'Rating']].head())
# actor_means
movies.sort_values(by=["Actor 1","Actor 2","Actor 3"],ascending=False).head()
```

```python
movies.sort_values(by=["Rating","Votes"],ascending=False).head(10)
# we can see there are times  when the votes are too less causing absurd ratings of 9+
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
genre_lists = movies['Genre'].apply(lambda x: x.split(', '))

pd.set_option('display.max_columns', None)

mlb = MultiLabelBinarizer()
genre_encoded = mlb.fit_transform(genre_lists)
# genre_encoded
genre_df = pd.DataFrame(genre_encoded, columns=mlb.classes_, index=movies.index).astype(bool)
genre_df
```

```python
genre_df["Music"]=genre_df["Music"] | genre_df["Musical"]
genre_df=genre_df.drop(["Musical"],axis=1)

(genre_df[genre_df["Music"]==1]).head()
```

```python

movies_final = pd.concat([movies, genre_df], axis=1).drop(['Genre','Name'], axis=1)

print(movies_final.head())
```

```python
movies_final.Votes=movies.Votes.str.replace(',', '').astype("Int64")
movies_final.dtypes
# Preprocessing complete!
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import mean_absolute_error

def evaluate_and_fit(df, model, target_col='Rating', n_splits=5):
    X = df.drop(columns=[target_col])
    y = df[target_col]
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=1)
    cv_scores = cross_val_score(model, X, y, cv=kf, scoring='neg_mean_absolute_error')
    mae_scores = -cv_scores
    avg_mae = np.mean(mae_scores)
    model.fit(X, y)
    print(f"Model: {type(model).__name__}")
    print(f"Average MAE ({n_splits} folds): {avg_mae:.4f}")
    print(f"MAE Std Dev: {np.std(mae_scores):.4f}")
    print("-" * 30)
    return model, avg_mae
```

```python
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor

xgb_model, xgb_mae = evaluate_and_fit(movies_final, XGBRegressor(n_estimators=100))
rf_model, rf_mae = evaluate_and_fit(movies_final, RandomForestRegressor(n_estimators=100))
```

```python
# XGboost model took the least time(a second or 2) while rf took around 20 seconds+
```
